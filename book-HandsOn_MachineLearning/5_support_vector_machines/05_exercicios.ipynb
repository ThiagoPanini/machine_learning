## Exercício 1

_What is the fundamental idea behind Support Vector Machines?_

R: A ideia fundamental por trás do algoritmo SVM é encontrar a maior "margem" entre as classes. Em outras palavras, o objetivo é definir uma fronteira de decisão que separa classes com margens que possuam maior distância com relação a esta fronteira de decisão.

## Exercício 2

_What is a support vector?_

R: Após treinado um modelo SVM, é nomeado vetor de suporte qualquer instância localizada na "rua", incluindo as bordas. A fronteira de decisão é inteiramente determinada pelos vetores de suporte e, qualquer instância que não seja um vetor de suporte, não possui qualquer influência no modelo. As predições também são calculadas considerando apenas os vetores de suporte, e não todo o conjunto de dados.

## Exercício 3

_Why is it important to scale the inputs when using SVMs?_

R: O algoritmo de SVM tende a encontrar a maior margem entre as classes e, se o conjunto de dados não está normalizado, features com valores baixos serão negligenciadas pelo modelo.

## Exercício 4

_Can an SVM classifier output a confidence score when it classifies an instance? What about the probability?_

R: Um classificador SVM pode retornar a distância entre uma instância de teste e a fronteira de decisão, e isto pode ser utilizado como um score de confiança. Contudo este score não pode ser diretamente convertido em probabilidade. Se ao instanciar uma classe de SVM for setadao o hiperparâmetro `probability=True`, as probabilidades serão calibradas utilizando Regressão Logística nos scores SVM obtidos. Isto irá adicionar os métodos `predict_proba()` e `predict_log_proba()` ao modelo.

## Exercício 5

_Should you use the primal or the dual form of the SVM problem to train a model ona  traning set with millions of instances and hundreds of features?_

R: Esta questão aplica-se apenas a SVMs lineares, uma vez que a aplicação do kernel restringe-se a forma dual. A complexidade computacional da forma primária é proporcional ao número de instâncias `m`. Já a complexidade da forma dual é proporcional a um número entre $m^2$ e $m^3$. Portanto, se há milhões de dados, a forma primária é preferível, uma vez que a forma dual apresentará grande lentidão.

### LinearSVC

```python
# Importando biblioteca e treinando modelo
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

# Criando pipeline
scaler = StandardScaler()
linear_svc = LinearSVC()
scaled_linear_svc = Pipeline([
    ('scaler', scaler),
    ('linear_svc', linear_svc)
])

# Treinando modelo
scaled_linear_svc.fit(X, y)
```

```python
def calculate_support_vectors(xlim, ylim, svm_clf):
    """docstring for calculateSupportVectors:
    função responsável por calcular e retornar as variáveis responsáveis pela plotagem
    dos vetores de suporte"""

    # create grid to evaluate model
    xx = np.linspace(xlim[0], xlim[1], 30)
    yy = np.linspace(ylim[0], ylim[1], 30)
    YY, XX = np.meshgrid(yy, xx)
    xy = np.vstack([XX.ravel(), YY.ravel()]).T
    Z = svm_clf.decision_function(xy).reshape(XX.shape)
    
    return XX, YY, Z
```

```python
# Função para plotagem da fronteira de decisão
def plot_decision_boundary(model1, model2=None, same_plot=False):
        
    # Plotando resultado
    fig, ax = plt.subplots()
    custom_colors = ["#3498db", "#C8391A"]
    ax = sns.scatterplot(x='x1', y='x2', data=df, hue='y', palette=custom_colors)

    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    XX, YY, Z = calculate_support_vectors(xlim, ylim, model1)
    
    if same_plot:
        XX2, YY2, Z2 = calculate_support_vectors(xlim, ylim, model2)
        ax.scatter(model1.support_vectors_[:, 0], model1.support_vectors_[:, 1], s=100,
                           linewidth=1, facecolors='none', edgecolors='k')
        ax.scatter(model2.support_vectors_[:, 0], model2.support_vectors_[:, 1], s=100,
                           linewidth=1, facecolors='none', edgecolors='k')
        ax.contour(XX, YY, Z, colors='k', levels=0, alpha=0.5,
                               linestyles='--')
        ax.contour(XX2, YY2, Z2, colors='b', levels=0, alpha=0.5,
                               linestyles='--')
        ax.set_title('Fronteiras de Decisão')
        plt.legend()
    else:    
        ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
                               linestyles=['--', '-', '--'])
        ax.scatter(model1.support_vectors_[:, 0], model1.support_vectors_[:, 1], s=100,
                               linewidth=1, facecolors='none', edgecolors='k')
    if not same_plot:
        try:
            model_kernel = model1.kernel
            ax.set_title(f'SVC(kernel="{model_kernel}", C={model1.C})')
        except:
            ax.set_title(f'LinearSVC(C={model1.C})')
    plt.show()
```

```python
# Definindo função para calcular vetores de suporte para LinearSVC
def linearSVC_support_vectors(model):
    # Convert to unscaled parameters
    b1 = model.decision_function([-scaler.mean_ / scaler.scale_])
    w1 = model.coef_[0] / scaler.scale_

    model.intercept_ = np.array([b1])
    model.coef_ = np.array([w1])

    # Find support vectors (LinearSVC does not do this automatically)
    t = y * 2 - 1
    support_vectors_idx1 = (t * (X.dot(w1) + b1) < 1).ravel()
    model.support_vectors_ = X[support_vectors_idx1]
```

```python
# PLotando resultado LinearSVC
linearSVC_support_vectors(linear_svc)
plot_decision_boundary(linear_svc)
```

### SVC

```python
# Importando bibliotecas
from sklearn.svm import SVC

svm_clf = SVC(kernel="linear")
svm_clf.fit(X, y)
```

```python
# Plotando resultado
plot_decision_boundary(svm_clf)
```

Comparando os resultados:

* **LinearSVC:** Com o hiperparâmetro C=1, a margem de classificação criada pelo modelo LinearSVC não foi tão ampla se comparada ao SVC com kernel linear. O outlier presente foi considerado para o cálculo da fronteira de decisão.

* **SVC:** Já o modelo SVC apresentou uma margem de classificação mais razoável, dado que desconsiderou o outlier presente.

```python
# Comparando os dois
plot_decision_boundary(linear_svc, svm_clf, same_plot=True)
```
