O algoritmo de SVM é uma maneira versátil e poderosa de se aplicar Machine Learning, sendo capaz de implementar, de modo linear ou não linear, modelos de classificação, regressão e até mesmo detecção de anomalia. Support Vector Machine pode ser considerado um dos mais populares algoritmos de Machine Learning, tendo o ápice de sua performance em classificação de dados complexos, porém de tamanho `pequeno` ou `médio`. 

## Linear SVM Classification

Retomando o dataset `iris` visto no capítulo anterior, a ideia por trás da classificação linear através do SVM pode ser explicada através dos gráficos abaixo.

```python
# Importando bibliotecas
from sklearn.svm import SVC
from sklearn import datasets

# Lendo dados
iris = datasets.load_iris()
X = iris["data"][:, (2, 3)]  # petal length, petal width
y = iris["target"]

# Indexando apenas duas das três classes (y == 2, virginica, fica de fora)
setosa_or_versicolor = (y == 0) | (y == 1)
X = X[setosa_or_versicolor]
y = y[setosa_or_versicolor]

# SVM Classifier model
svm_clf = SVC(kernel="linear", C=float("inf"))
svm_clf.fit(X, y)
```

```python
# Bad models
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x0 = np.linspace(0, 5.5, 200)
pred_1 = 5*x0 - 20
pred_2 = x0 - 1.8
pred_3 = 0.1 * x0 + 0.5

def plot_svc_decision_boundary(svm_clf, xmin, xmax):
    w = svm_clf.coef_[0]
    b = svm_clf.intercept_[0]

    # At the decision boundary, w0*x0 + w1*x1 + b = 0
    # => x1 = -w0/w1 * x0 - b/w1
    x0 = np.linspace(xmin, xmax, 200)
    decision_boundary = -w[0]/w[1] * x0 - b/w[1]

    margin = 1/w[1]
    gutter_up = decision_boundary + margin
    gutter_down = decision_boundary - margin

    svs = svm_clf.support_vectors_
    plt.scatter(svs[:, 0], svs[:, 1], s=180, facecolors='#FFAAAA')
    plt.plot(x0, decision_boundary, "k-", linewidth=2)
    plt.plot(x0, gutter_up, "k--", linewidth=2)
    plt.plot(x0, gutter_down, "k--", linewidth=2)

plt.figure(figsize=(12,4))

plt.subplot(121)
plt.plot(x0, pred_1, "g--", linewidth=2)
plt.plot(x0, pred_2, "m-", linewidth=2)
plt.plot(x0, pred_3, "r-", linewidth=2)
plt.plot(X[:, 0][y==1], X[:, 1][y==1], "bs", label="Iris-Versicolor")
plt.plot(X[:, 0][y==0], X[:, 1][y==0], "yo", label="Iris-Setosa")
plt.xlabel("Petal length", fontsize=14)
plt.ylabel("Petal width", fontsize=14)
plt.legend(loc="upper left", fontsize=14)
plt.axis([0, 5.5, 0, 2])

plt.subplot(122)
plot_svc_decision_boundary(svm_clf, 0, 5.5)
plt.plot(X[:, 0][y==1], X[:, 1][y==1], "bs")
plt.plot(X[:, 0][y==0], X[:, 1][y==0], "yo")
plt.xlabel("Petal length", fontsize=14)
plt.axis([0, 5.5, 0, 2])

plt.show()
```

As duas classes presentes na ilustração acima (Iris-Versicolor e Iris-Setosa) são linearmente separáveis. O gráfico a esquerda mostra fronteiras de decisão de três possíveis classificadores. O modelo representado pela linha verde tracejada possui uma performance tão baixa que mal consegue realizar a separação entre as classes. Os outros dois modelos, representados pelas linhas vermelha e roxa, são capazes de separar linearmente as classes, entretanto, a fronteira de decisão resultante é tão próxima das amostras que provavelmente os modelos não terão bons resultados ao serem apresentados a novas instâncias.

Em contraste, a fronteira de decisão do gráfico a direita representa um modelo treinado com o algoritmo SVM que, por sua vez, não apenas realiza a separação das classes, como também mantém a maior distância possívei das instâncias. É possível pensar em um `classificador SVM` como uma forma de encontrar a "rua" mais larga possível (representada pelas linhas tracejadas paralelas). Isto é conhecido como _large margin classification_

**Importante:** notem que a adição de exemplos "fora" da rua (ou fora da margem) não afeta a fronteira de decisão, uma vez que esta é determinada pelas instâncias presentes nas margens da classificação. Estas instâncias são conhecidas como `support vectors` (ou vetores de suporte) e estão identificados com um círculo.

```python
# Sensibilização a normalização dos dados
Xs = np.array([[1, 50], [5, 20], [3, 80], [5, 60]]).astype(np.float64)
ys = np.array([0, 0, 1, 1])
svm_clf = SVC(kernel="linear", C=100)
svm_clf.fit(Xs, ys)

plt.figure(figsize=(12,3.2))
plt.subplot(121)
plt.plot(Xs[:, 0][ys==1], Xs[:, 1][ys==1], "bo")
plt.plot(Xs[:, 0][ys==0], Xs[:, 1][ys==0], "ms")
plot_svc_decision_boundary(svm_clf, 0, 6)
plt.xlabel("$x_0$", fontsize=20)
plt.ylabel("$x_1$  ", fontsize=20, rotation=0)
plt.title("Unscaled", fontsize=16)
plt.axis([0, 6, 0, 90])

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(Xs)
svm_clf.fit(X_scaled, ys)

plt.subplot(122)
plt.plot(X_scaled[:, 0][ys==1], X_scaled[:, 1][ys==1], "bo")
plt.plot(X_scaled[:, 0][ys==0], X_scaled[:, 1][ys==0], "ms")
plot_svc_decision_boundary(svm_clf, -2, 2)
plt.xlabel("$x_0$", fontsize=20)
plt.title("Scaled", fontsize=16)
plt.axis([-2, 2, -2, 2]);
```

A comparação acima demonstra os efeitos da `normalização` dos dados em um algoritmo de classificação linear por SVM.

### Soft Margin Classification

Suponhamos um espalhamento dos dados onde todos encontram-se do lado direito da "rua". A isso chamamos _hard margin classification_, trazendo consigo dois principais problemas:
* Funciona apenas quando os dados são linearmente separáveis;
* Há uma grande sensibilidade a outliers.

Vejamos um exemplo:

```python
X_outliers = np.array([[3.4, 1.3], [3.2, 0.8]])
y_outliers = np.array([0, 0])
Xo1 = np.concatenate([X, X_outliers[:1]], axis=0)
yo1 = np.concatenate([y, y_outliers[:1]], axis=0)
Xo2 = np.concatenate([X, X_outliers[1:]], axis=0)
yo2 = np.concatenate([y, y_outliers[1:]], axis=0)

svm_clf2 = SVC(kernel="linear", C=10**9)
svm_clf2.fit(Xo2, yo2)

plt.figure(figsize=(12,2.7))

plt.subplot(121)
plt.plot(Xo1[:, 0][yo1==1], Xo1[:, 1][yo1==1], "bs")
plt.plot(Xo1[:, 0][yo1==0], Xo1[:, 1][yo1==0], "yo")
plt.text(0.3, 1.0, "Impossible!", fontsize=24, color="red")
plt.xlabel("Petal length", fontsize=14)
plt.ylabel("Petal width", fontsize=14)
plt.annotate("Outlier",
             xy=(X_outliers[0][0], X_outliers[0][1]),
             xytext=(2.5, 1.7),
             ha="center",
             arrowprops=dict(facecolor='black', shrink=0.1),
             fontsize=16,
            )
plt.axis([0, 5.5, 0, 2])

plt.subplot(122)
plt.plot(Xo2[:, 0][yo2==1], Xo2[:, 1][yo2==1], "bs")
plt.plot(Xo2[:, 0][yo2==0], Xo2[:, 1][yo2==0], "yo")
plot_svc_decision_boundary(svm_clf2, 0, 5.5)
plt.xlabel("Petal length", fontsize=14)
plt.annotate("Outlier",
             xy=(X_outliers[1][0], X_outliers[1][1]),
             xytext=(3.2, 0.08),
             ha="center",
             arrowprops=dict(facecolor='black', shrink=0.1),
             fontsize=16,
            )
plt.axis([0, 5.5, 0, 2])

plt.show()
```

Percebe-se que, a esquerda, a presença do outlier inviabiliza a construção de uma fronteira de decisão que realize a total separação dos dados. No gráfico a direita, foi possível separar os dados linearmente, entretanto a presença do outlier causou tal efeito na fronteira de decisão que esta ficou completamente aquém do esperado.

Como obter um modelo mais `flexível` para evitar este tipo de comportamento?

Para encontrar um bom balanceamento no modelo, isto é, manter a margem de classificação tão grande quanto possível, limitando os limites de violação (instâncias que permanecem no meio da margem ou do lado errado da classificação), precisamos de uma nova abordagem. A ela chamamos de _soft margin classification_. Para controlar este balanço, utilizamos o hyperparâmetro `C`.

* C pode ser considerado o inverso do parâmetro de regularização $\alpha$ ou $\lambda$;
* Quanto menor C, maior a margem de classificação, porém maior o índice de violação de margem;
* Quanto maior C, menor a margem de classificação e menor a quantidade de violação de margem.

```python
import numpy as np
from sklearn import datasets
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC

iris = datasets.load_iris()
X = iris["data"][:, (2, 3)]  # petal length, petal width
y = (iris["target"] == 2).astype(np.float64)  # Iris-Virginica

svm_clf = Pipeline([
        ("scaler", StandardScaler()),
        ("linear_svc", LinearSVC(C=1, loss="hinge", random_state=42)),
    ])

svm_clf.fit(X, y)
```

Gerando o gráfico com diferentes índices de regularização C

```python
scaler = StandardScaler()
svm_clf1 = LinearSVC(C=1, loss="hinge", random_state=42)
svm_clf2 = LinearSVC(C=100, loss="hinge", random_state=42)

scaled_svm_clf1 = Pipeline([
        ("scaler", scaler),
        ("linear_svc", svm_clf1),
    ])
scaled_svm_clf2 = Pipeline([
        ("scaler", scaler),
        ("linear_svc", svm_clf2),
    ])

scaled_svm_clf1.fit(X, y)
scaled_svm_clf2.fit(X, y)
```

```python
# Convert to unscaled parameters
b1 = svm_clf1.decision_function([-scaler.mean_ / scaler.scale_])
b2 = svm_clf2.decision_function([-scaler.mean_ / scaler.scale_])
w1 = svm_clf1.coef_[0] / scaler.scale_
w2 = svm_clf2.coef_[0] / scaler.scale_
svm_clf1.intercept_ = np.array([b1])
svm_clf2.intercept_ = np.array([b2])
svm_clf1.coef_ = np.array([w1])
svm_clf2.coef_ = np.array([w2])

# Find support vectors (LinearSVC does not do this automatically)
t = y * 2 - 1
support_vectors_idx1 = (t * (X.dot(w1) + b1) < 1).ravel()
support_vectors_idx2 = (t * (X.dot(w2) + b2) < 1).ravel()
svm_clf1.support_vectors_ = X[support_vectors_idx1]
svm_clf2.support_vectors_ = X[support_vectors_idx2]
```

```python
plt.figure(figsize=(12,3.2))
plt.subplot(121)
plt.plot(X[:, 0][y==1], X[:, 1][y==1], "g^", label="Iris-Virginica")
plt.plot(X[:, 0][y==0], X[:, 1][y==0], "bs", label="Iris-Versicolor")
plot_svc_decision_boundary(svm_clf1, 4, 6)
plt.xlabel("Petal length", fontsize=14)
plt.ylabel("Petal width", fontsize=14)
plt.legend(loc="upper left", fontsize=14)
plt.title("$C = {}$".format(svm_clf1.C), fontsize=16)
plt.axis([4, 6, 0.8, 2.8])

plt.subplot(122)
plt.plot(X[:, 0][y==1], X[:, 1][y==1], "g^")
plt.plot(X[:, 0][y==0], X[:, 1][y==0], "bs")
plot_svc_decision_boundary(svm_clf2, 4, 6)
plt.xlabel("Petal length", fontsize=14)
plt.title("$C = {}$".format(svm_clf2.C), fontsize=16)
plt.axis([4, 6, 0.8, 2.8]);
```

Se o modelo sofre de `overfitting`, é possível reduzir este efeito através da redução do hiperparâmetro `C`.

Diferente do modelo de `Regressão Logística`, o modelo SVM não retorna como output a probabilidade de determinada instância pertencer a determinada classe.

Uma outra alternativa para os testes apresentados acima seria utilizar a classe SVC com as configurações
    
    SVC(kernel="linear", C=1)

Porém assim o treinamento rodaria mais devagar e, especialmente em grandes conjutos de dados, isto não é recomendado.

## Nonlinear SVM Classificação

Apesar do modelo linear de classificação com SVM ser eficiente e performar surpreendentemente bem em muitos casos, muitos datasets não estão nem um pouco próximos de serem linearmente separáveis. Uma abordagem possível refere-se a adição de novas features, como features polinomiais, por exemplo. Em alguns casos, este procedimento pode transformar dados não linearmente separáveis em dados linearmente separáveis. Vejamos um exemplo prático:

```python
X1D = np.linspace(-4, 4, 9).reshape(-1, 1)
X2D = np.c_[X1D, X1D**2]
y = np.array([0, 0, 1, 1, 1, 1, 1, 0, 0])

plt.figure(figsize=(11, 4))

plt.subplot(121)
plt.grid(True, which='both')
plt.axhline(y=0, color='k')
plt.plot(X1D[:, 0][y==0], np.zeros(4), "bs")
plt.plot(X1D[:, 0][y==1], np.zeros(5), "g^")
plt.gca().get_yaxis().set_ticks([])
plt.xlabel(r"$x_1$", fontsize=20)
plt.axis([-4.5, 4.5, -0.2, 0.2])
plt.title('Nonlinear Data')

plt.subplot(122)
plt.grid(True, which='both')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.plot(X2D[:, 0][y==0], X2D[:, 1][y==0], "bs")
plt.plot(X2D[:, 0][y==1], X2D[:, 1][y==1], "g^")
plt.xlabel(r"$x_1$", fontsize=20)
plt.ylabel(r"$x_2$", fontsize=20, rotation=0)
plt.gca().get_yaxis().set_ticks([0, 4, 8, 12, 16])
plt.plot([-4.5, 4.5], [6.5, 6.5], "r--", linewidth=3, label='dec bound')
plt.axis([-4.5, 4.5, -1, 17])
plt.title('Linearly separable')
plt.legend()

plt.subplots_adjust(right=1)

plt.show()
```

Um outro exemplo pode ser evidenciado através de um pipeline de treinamento para classificar o dataset `make moons`.

```python
# Dados dataset make moons
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=100, noise=0.15, random_state=42)

def plot_dataset(X, y, axes):
    plt.plot(X[:, 0][y==0], X[:, 1][y==0], "bs")
    plt.plot(X[:, 0][y==1], X[:, 1][y==1], "g^")
    plt.axis(axes)
    plt.grid(True, which='both')
    plt.xlabel(r"$x_1$", fontsize=20)
    plt.ylabel(r"$x_2$", fontsize=20, rotation=0)

plot_dataset(X, y, [-1.5, 2.5, -1, 1.5])
plt.show()
```

```python
# Treinando modelo
from sklearn.datasets import make_moons
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures

polynomial_svm_clf = Pipeline([
        ("poly_features", PolynomialFeatures(degree=3)),
        ("scaler", StandardScaler()),
        ("svm_clf", LinearSVC(C=10, loss="hinge", random_state=42))
    ])

polynomial_svm_clf.fit(X, y)
```

```python
# Fronteira de decisão
def plot_predictions(clf, axes):
    x0s = np.linspace(axes[0], axes[1], 100)
    x1s = np.linspace(axes[2], axes[3], 100)
    x0, x1 = np.meshgrid(x0s, x1s)
    X = np.c_[x0.ravel(), x1.ravel()]
    y_pred = clf.predict(X).reshape(x0.shape)
    y_decision = clf.decision_function(X).reshape(x0.shape)
    plt.contourf(x0, x1, y_pred, cmap=plt.cm.brg, alpha=0.2)
    plt.contourf(x0, x1, y_decision, cmap=plt.cm.brg, alpha=0.1)

plot_predictions(polynomial_svm_clf, [-1.5, 2.5, -1, 1.5])
plot_dataset(X, y, [-1.5, 2.5, -1, 1.5])

plt.show()
```

### Polynomial Kernel

Adicionar features polinomiais é um procedimento de fácil execução e funciona bem com qualquer algoritmo de Machine Learning. Entretanto, dois pontos de dificuldade precisam ser pontuados sobre esta abordagem:
* **1.** Em graus polinomiais baixo, o modelo não consegue tratar datasets complexos;
* **2.** Em graus polinomiais elevados, é necessário lidar com uma grande quantidade de features adicionais, diminuindo a velocidade de execução do modelo.

Felizmente, ao usar SVM, é possível aplicar um procedimeneto matemático de grande utilidade chamado `kernel trick`. Com isto, é possível alcançar os mesmos resultados trazidos por uma adição de features polinomiais em alto grau, sem a necessidade de adiciona-las de fato. Não há um aumento drástico no número de features! Vamos testar no dataset makemoons.

```python
from sklearn.svm import SVC

poly_kernel_svm_clf = Pipeline([
        ("scaler", StandardScaler()),
        ("svm_clf", SVC(kernel="poly", degree=3, coef0=1, C=5))
    ])
poly_kernel_svm_clf.fit(X, y)
```

```python
# Treinando modelo
poly100_kernel_svm_clf = Pipeline([
        ("scaler", StandardScaler()),
        ("svm_clf", SVC(kernel="poly", degree=10, coef0=100, C=5))
    ])
poly100_kernel_svm_clf.fit(X, y)
```

```python
# Verificando resultado
plt.figure(figsize=(11, 4))

plt.subplot(121)
plot_predictions(poly_kernel_svm_clf, [-1.5, 2.5, -1, 1.5])
plot_dataset(X, y, [-1.5, 2.5, -1, 1.5])
plt.title(r"$d=3, r=1, C=5$", fontsize=18)

plt.subplot(122)
plot_predictions(poly100_kernel_svm_clf, [-1.5, 2.5, -1, 1.5])
plot_dataset(X, y, [-1.5, 2.5, -1, 1.5])
plt.title(r"$d=10, r=100, C=5$", fontsize=18)

plt.show()
```

A esquerda, temos um modelo treinado com um grau polinomial `d=3`. A direita, temos um modelo com `d=10`. Se o modelo sofre de overfitting, podemos aumentar o hiperparâmetro `d` para simular curvas de maior grau polinomial. O contrário também é verdadeiro para casos de overfitting.

O hiperparâmetro `coef0` ou `r`, pelos gráficos, controla o quanto o modelo será influenciado pelo grau polinomial.

### Adding Similarity Features

Uma outra técnica para se trabalhar com problemas não lineares é adicionar features calculadas através de uma `função de similaridade`, sendo esta responsável por medir quanto cada instância se assemelha a um `ponto de referência` particular (_landmark_)

Gaussian Radial Basis Function (RBF)

\begin{equation*}
    \phi_y(x, l)=\text{exp } \left(-y\mid\mid{x-l}\mid\mid^2\right)
\end{equation*}

Esta é uma função em formato de sino (bell-shaped) variando entre 0 (_muito longe do ponto de referência, ou landmark_) até 1 (_exatamente no ponto de referência ou landmark_). Agora estamos prontos para calcular as novas features.

```python
def gaussian_rbf(x, landmark, gamma):
    return np.exp(-gamma * np.linalg.norm(x - landmark, axis=1)**2)

gamma = 0.3

x1s = np.linspace(-4.5, 4.5, 200).reshape(-1, 1)
x2s = gaussian_rbf(x1s, -2, gamma)
x3s = gaussian_rbf(x1s, 1, gamma)

XK = np.c_[gaussian_rbf(X1D, -2, gamma), gaussian_rbf(X1D, 1, gamma)]
yk = np.array([0, 0, 1, 1, 1, 1, 1, 0, 0])

plt.figure(figsize=(11, 4))

plt.subplot(121)
plt.grid(True, which='both')
plt.axhline(y=0, color='k')
plt.scatter(x=[-2, 1], y=[0, 0], s=150, alpha=0.5, c="red")
plt.plot(X1D[:, 0][yk==0], np.zeros(4), "bs")
plt.plot(X1D[:, 0][yk==1], np.zeros(5), "g^")
plt.plot(x1s, x2s, "g--")
plt.plot(x1s, x3s, "b:")
plt.gca().get_yaxis().set_ticks([0, 0.25, 0.5, 0.75, 1])
plt.xlabel(r"$x_1$", fontsize=20)
plt.ylabel(r"Similarity", fontsize=14)
plt.annotate(r'$\mathbf{x}$',
             xy=(X1D[3, 0], 0),
             xytext=(-0.5, 0.20),
             ha="center",
             arrowprops=dict(facecolor='black', shrink=0.1),
             fontsize=18,
            )
plt.text(-2, 0.9, "$x_2$", ha="center", fontsize=20)
plt.text(1, 0.9, "$x_3$", ha="center", fontsize=20)
plt.axis([-4.5, 4.5, -0.1, 1.1])

plt.subplot(122)
plt.grid(True, which='both')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.plot(XK[:, 0][yk==0], XK[:, 1][yk==0], "bs")
plt.plot(XK[:, 0][yk==1], XK[:, 1][yk==1], "g^")
plt.xlabel(r"$x_2$", fontsize=20)
plt.ylabel(r"$x_3$  ", fontsize=20, rotation=0)
plt.annotate(r'$\phi\left(\mathbf{x}\right)$',
             xy=(XK[3, 0], XK[3, 1]),
             xytext=(0.65, 0.50),
             ha="center",
             arrowprops=dict(facecolor='black', shrink=0.1),
             fontsize=18,
            )
plt.plot([-0.1, 1.1], [0.57, -0.1], "r--", linewidth=3)
plt.axis([-0.1, 1.1, -0.1, 1.1])
    
plt.subplots_adjust(right=1)

plt.show()
```

No gráfico a esquerda, dois _landmarks_ são posicionados nos pontos:
* $x=-2$ e 
* $x=1$

Temos nosso ponto de exemplo localizado em  **$x=-1$**

Sabemos que as distâncias do nosso ponto **x** para os _landmarks_ são:
* 1 para o primeiro landmark;
* 2 para o segundo landmark;

Dessa forma, temos as novas features $x_2$ e $x_3$ como sendo:

\begin{equation*}
    x_2=(-0.3\times1^2)\approx0.74
\end{equation*}
\begin{equation*}
    x_3=(-0.3\times2^2)\approx0.30
\end{equation*}

Temos então este novo ponto indicado no gráfico a direia, assim como todos os outros pontos. Os dados agora são `linearmente separáveis`.

```python
x1_example = X1D[3, 0]
for landmark in (-2, 1):
    k = gaussian_rbf(np.array([[x1_example]]), np.array([[landmark]]), gamma)
    print("Phi({}, {}) = {}".format(x1_example, landmark, k))
```

Pergunta principal:
    - Onde colocar os pontos de referência?

Resposta simples:
    - Os landmarks podem ser colocados no local de cada uma das instâncias, gerando uma nova feature para cada amostra, aumentando as chances dos dados transformados serem linearmente separáveis.

Ponto negativo:
    - Teremos um dataset com m instâncias e m features

### Gaussian RBF Kernel

Novamente, semalhante ao procedimento de adição de features polinomiais, a aplicação das funções de similaridade funcionam bem com qualquer algoritmo de Machine Learning, tendo também o viés computacional devido a adição de muitas features.

Da mesma forma, o modelo SVM magicamente auxilia nesse ponto através de mais um `kernel trick`.

```python
rbf_kernel_svm_clf = Pipeline([
        ("scaler", StandardScaler()),
        ("svm_clf", SVC(kernel="rbf", gamma=5, C=0.001))
    ])
rbf_kernel_svm_clf.fit(X, y)
```

```python
from sklearn.svm import SVC

gamma1, gamma2 = 0.1, 5
C1, C2 = 0.001, 1000
hyperparams = (gamma1, C1), (gamma1, C2), (gamma2, C1), (gamma2, C2)

svm_clfs = []
for gamma, C in hyperparams:
    rbf_kernel_svm_clf = Pipeline([
            ("scaler", StandardScaler()),
            ("svm_clf", SVC(kernel="rbf", gamma=gamma, C=C))
        ])
    rbf_kernel_svm_clf.fit(X, y)
    svm_clfs.append(rbf_kernel_svm_clf)

plt.figure(figsize=(11, 7))

for i, svm_clf in enumerate(svm_clfs):
    plt.subplot(221 + i)
    plot_predictions(svm_clf, [-1.5, 2.5, -1, 1.5])
    plot_dataset(X, y, [-1.5, 2.5, -1, 1.5])
    gamma, C = hyperparams[i]
    plt.title(r"$\gamma = {}, C = {}$".format(gamma, C), fontsize=16)

plt.tight_layout()
plt.show()
```

Acima temos uma comparação de combinação entre hiperparâmetros. Aumentar o hiperparâmetro $gamma$ ($\gamma$) deixa a curva de sino (bell-shaped) mais estreita (verificar plot anterior) e, como resultado, o range de influência de cada instância é menor: a fronteira de decisão acaba sendo mais irregular, oscilando ao redor de instâncias individuais.

Por outro lado, um valor baixo para $gamma$ ($\gamma$) aumenta o shape da curva de sino, fazendo com que as instâncias possuam um grande range de influência: a fronteira de decisão se torna mais suave.

Dessa forma, $\gamma$ funciona como um parâmetro de `regularização`:
* No caso de overfitting, a recomendação é reduzir gamma
* No caso de underfitting, uma tentativa válida é aumentar gamma

Similar ao hiperparâmetro `C`.

Existem muitos outros `kernels`, alguns para aplicações bem específicas. Qual utilizar?
**Recomendação:**
    * Tentar linear kernel em primeiro lugar
        * Lembrar que LinearSVC é muito mais rápido que SVC(kernel='linear')
        * Especialmente se há muitos dados ou muitas features no dataset
    * Se o dataset não é tão grande, tentar Gaussian RBF kernel
    * Se houver poder computacional, tentar outros kernels com grid search e cross-validation

## SVM Regression

Como mencionado anteriormente, o algoritmo de SVM é extremamente versátil: não apenas suporta classificação linear e não-linear, como também suporta regressão linear e não-linear. O artifício da regressão é inverso ao da classificação: ao invés de tentar enquadrar a maior margem possível entre duas classes enquanto limita violações de margem, `SVM Regression` tenta enquadrar o `maior` número de instâncias possível `dentro` da margem, também limitando violações.

A largura da margem é controlada pelo hiperparâmtro $\epsilon$

```python
# Simulando dados
np.random.seed(42)
m = 50
X = 2 * np.random.rand(m, 1)
y = (4 + 3 * X + np.random.randn(m, 1)).ravel()
```

```python
from sklearn.svm import LinearSVR

svm_reg = LinearSVR(epsilon=1.5, random_state=42)
svm_reg.fit(X, y)
```

```python
svm_reg1 = LinearSVR(epsilon=1.5, random_state=42)
svm_reg2 = LinearSVR(epsilon=0.5, random_state=42)
svm_reg1.fit(X, y)
svm_reg2.fit(X, y)

def find_support_vectors(svm_reg, X, y):
    y_pred = svm_reg.predict(X)
    off_margin = (np.abs(y - y_pred) >= svm_reg.epsilon)
    return np.argwhere(off_margin)

svm_reg1.support_ = find_support_vectors(svm_reg1, X, y)
svm_reg2.support_ = find_support_vectors(svm_reg2, X, y)

eps_x1 = 1
eps_y_pred = svm_reg1.predict([[eps_x1]])
```

```python
def plot_svm_regression(svm_reg, X, y, axes):
    x1s = np.linspace(axes[0], axes[1], 100).reshape(100, 1)
    y_pred = svm_reg.predict(x1s)
    plt.plot(x1s, y_pred, "k-", linewidth=2, label=r"$\hat{y}$")
    plt.plot(x1s, y_pred + svm_reg.epsilon, "k--")
    plt.plot(x1s, y_pred - svm_reg.epsilon, "k--")
    plt.scatter(X[svm_reg.support_], y[svm_reg.support_], s=180, facecolors='#FFAAAA')
    plt.plot(X, y, "bo")
    plt.xlabel(r"$x_1$", fontsize=18)
    plt.legend(loc="upper left", fontsize=18)
    plt.axis(axes)

plt.figure(figsize=(9, 4))
plt.subplot(121)
plot_svm_regression(svm_reg1, X, y, [0, 2, 3, 11])
plt.title(r"$\epsilon = {}$".format(svm_reg1.epsilon), fontsize=18)
plt.ylabel(r"$y$", fontsize=18, rotation=0)
#plt.plot([eps_x1, eps_x1], [eps_y_pred, eps_y_pred - svm_reg1.epsilon], "k-", linewidth=2)
plt.annotate(
        '', xy=(eps_x1, eps_y_pred), xycoords='data',
        xytext=(eps_x1, eps_y_pred - svm_reg1.epsilon),
        textcoords='data', arrowprops={'arrowstyle': '<->', 'linewidth': 1.5}
    )
plt.text(0.91, 5.6, r"$\epsilon$", fontsize=20)
plt.subplot(122)
plot_svm_regression(svm_reg2, X, y, [0, 2, 3, 11])
plt.title(r"$\epsilon = {}$".format(svm_reg2.epsilon), fontsize=18)

plt.show()
```

Interessasnte citar que adicionar amostras `dentro` da margem não afeta o modelo, sendo este cosiderado como $\epsilon$ _sensitive_. Para plotar os gráficos acima, utilizamos o algoritmo `LinearSVR`.

Para abordar problemas não lineares de regressão, podemos utilizar o funcionamento do `kernel` do SVM. Vejamos um exemplo:

```python
# SVR não linear, kernel=poly
from sklearn.svm import SVR

np.random.seed(42)
m = 100
X = 2 * np.random.rand(m, 1) - 1
y = (0.2 + 0.1 * X + 0.5 * X**2 + np.random.randn(m, 1)/10).ravel()

svm_poly_reg = SVR(kernel="poly", degree=2, C=100, epsilon=0.1, gamma="auto")
svm_poly_reg.fit(X, y)
```

```python
svm_poly_reg1 = SVR(kernel="poly", degree=2, C=100, epsilon=0.1, gamma="auto")
svm_poly_reg2 = SVR(kernel="poly", degree=2, C=0.01, epsilon=0.1, gamma="auto")
svm_poly_reg1.fit(X, y)
svm_poly_reg2.fit(X, y)
```

```python
plt.figure(figsize=(9, 4))
plt.subplot(121)
plot_svm_regression(svm_poly_reg1, X, y, [-1, 1, 0, 1])
plt.title(r"$degree={}, C={}, \epsilon = {}$".format(svm_poly_reg1.degree, 
                                                     svm_poly_reg1.C, svm_poly_reg1.epsilon), 
          fontsize=18)
plt.ylabel(r"$y$", fontsize=18, rotation=0)
plt.subplot(122)
plot_svm_regression(svm_poly_reg2, X, y, [-1, 1, 0, 1])
plt.title(r"$degree={}, C={}, \epsilon = {}$".format(svm_poly_reg2.degree, 
                                                     svm_poly_reg2.C, svm_poly_reg2.epsilon), 
          fontsize=18)
plt.show()
```
