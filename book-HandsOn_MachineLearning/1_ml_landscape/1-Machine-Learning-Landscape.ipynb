<font size=4> **Tom Mitchell, 1977** - _É dito que um programa de computador aprende com a experiência E a respeito de uma tarefa T com uma performance P, se sua performance em T, medida por P, melhora através da experiência E_.</font>

Um exemplo clássico de primeiro algorítimo de Machine Learning bem sucedido (datado de 1990) diz respeito ao classificador de **_Spam_**. Aplicando a definição de Tom Mitchel no referido exemplo, tem-se que a tarefa T é a classificação de novos e-mails entre "spam" ou "não-spam". A experiência E são os dados de treinamento. Já a performance P é aquela definida pelo usuário, onde, por exemplo, é estabelecida uma acurácia de 95%.

<font size=4> **Por que usar Machine Learning?**</font>

Mais uma vez retomando o exemplo do classificador de **_Spam_**, é válido dizer que os programas maliciosos responsáveis pela criação de tais e-mails tendem a ligeiramente modificar o conteúdo das mensagens para que estas sejam cada vez mais difíceis de serem identificadas. Nuances como "4U" e "For You", por exemplo, são utilizadas em larga escala, o que dificulta um algorítimo treinado especificamente para identificar uma das duas variações, ou então palavras fixas como "credit card", "free", etc.

Não é eficiente ter que criar ou alterar os modelos cada vez que um novo *padrão* é encontrado.

Um algoritímo baseado nas técnicas de Machine Learning automaticamente *aprende* palavras e frases preditoras de spam através da detecção de padrões não-usuais dentro do conjunto de spams. Tudo isso sem a intervenção do programador. Parece mágica.

<font size=4> **Tipos de Sistemas Inteligentes**</font>

* <font size=3>**_Supervised Learning:_**</font> Algorítimo treinado através do fornecimento de inputs e **outputs** dentro de um conjunto de dados, ou seja, o treinamento é baseado em uma resposta conhecida. Alguns exemplos clássicos de aprendizagem supervisionada: **regressão** e **classificação**. Para treinar estes algorítimos, é necessário fornecer os elementos preditores (_features_) e também as soluções (_labels_). São exemplos de algorítimos:

    - k-Nearest Neighbors
    - Linear Regression
    - Logistic Regression
    - Support Vector Machines (SVMs)
    - Decision Trees and Random Forest
    - Neural networks

* <font size=3>**_Unsupervised Learning:_**</font> Neste modelo de aprendizagem, os _labels_ (soluções) não estão bem definidos. Técnicas de _clustering_ estão altamente associadas ao aprendizado não supervisionado pois, dado um conjunto de dados, o algorítimo é capaz de identificar e agrupar padrões, sem que sejam fornecidos dados prévios e tampouco o nome desses grupos. São exemplos de algorítimos de aprendizagem não supervisionada:

    - Clustering
        - k-Means
        - Hierarchical Cluster Analysis (HCA)
        - Expectation Maximization
    - Visualization and dimensionality reduction*
        - Principal Component Analysis (PCA)
        - Kernel PCA
        - Locally-Linear Embedding (LLE)
        - t-distributed Stochastic Neighbor Embedding (t-SNE)
    - Association rule learning
        - Apriori
        - Eclat

_*Parênteses sobre Redução de Dimensionalidade: está é a técnica que descreve a simplificação dos dados sem muitas perdas. Uma maneira de faze-lo é reunindo diersas features correlacionadas em uma única feature (feature extraction). Também é uma boa ideia aplicar técnicas de redução de dimensionalidade nos dados antes de realizar o treinamento, qualquer que seja o modelo utilizado e mesmo que este seja de aprendizagem supervisionada. Com isso, o modelo será treinado mais rapido, uma quantidade menor de memória será utilizada e, em alguns casos, o modelo irá performar melhor._

* <font size=3>**_Semisupervised Learning:_**</font> Algorítimos de aprendizagem semi-supervisionada onde as soluções são dadas parcialmente, deixando a cargo do próprio algorítimo, identficações e correlações futuras. Um bom exemplo disso é o Google Photos, onde o algorítimo traz, com certa eficiência, rostos semelhantes (clustering não supervisionado), sendo somente necessário dizer à ele a quem pertence a foto (output label supervisionado).

* <font size=3>**_Reinforcement Learning:_**</font> O aprendizado por reforço é o que de mais diferente se apresenta dentro dos algorítimos de Machine Learning. Nele, o sistema de aprendizado (chamado de _agente_), consegue observar o ambiente, selecionar e performar ações, coletando _recompensas_ ou _penalidades_, de acordo com as ações tomadas. Por conta própria, através do retorno obtido, o modelo aprende através de uma estratégia, chamada _policy_, cujo objetivo é coletar o maior número de _recompensas_. Exemplos:

    - Programa alphaGo da DeepMind.

```python
# Função para preparar os dados (TENTATIVA )
def prepare_country_stats(df_bli, df_gdp):
    country = []
    bli = []
    gdp = []
    df_gdp_country = df_gdp['Country']
    for c in df_gdp_country:
        if c.upper().strip() in df_bli['Country'].apply(lambda x: x.upper().strip()).values:
            country.append(c)
            bli.append(df_bli.loc[lambda df: np.logical_and(np.logical_and(df['Country']==c, df['Unit']=='Percentage'),
                                         df['Value'] < 40), 'Value'].mean())
            gdp.append(df_gdp.loc[lambda df: df['Country']==c, '2015'].values[0])
    dictio = {'country': country, 'gdp': gdp, 'life_satisfaction': bli}
    return pd.DataFrame(dictio)
```

```python
# Preparação dos datasets
def prepare_country_stats(oecd_bli, gdp_per_capita):
    oecd_bli = oecd_bli[oecd_bli["INEQUALITY"]=="TOT"]
    oecd_bli = oecd_bli.pivot(index="Country", columns="Indicator", values="Value")
    gdp_per_capita.rename(columns={"2015": "GDP per capita"}, inplace=True)
    gdp_per_capita.set_index("Country", inplace=True)
    full_country_stats = pd.merge(left=oecd_bli, right=gdp_per_capita,
                                  left_index=True, right_index=True)
    full_country_stats.sort_values(by="GDP per capita", inplace=True)
    remove_indices = [0, 1, 6, 8, 33, 34, 35]
    keep_indices = list(set(range(36)) - set(remove_indices))
    return full_country_stats[["GDP per capita", 'Life satisfaction']].iloc[keep_indices]
```

```python
# Code example
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn.linear_model

# Load data
df_bli = pd.read_csv('better_life_index.csv', thousands=',')
df_gdp = pd.read_csv('gdp_per_capita.csv', 
                     thousands=',', delimiter='\t', encoding='latin1', na_values='n/a')

# Prepare the data
country_stats = prepare_country_stats(df_bli, df_gdp)
X = np.c_[country_stats["GDP per capita"]]
y = np.c_[country_stats["Life satisfaction"]]

# Visualize the data
country_stats.plot(kind='scatter', x="GDP per capita", y='Life satisfaction')
plt.show()

# Select a linear model
model = sklearn.linear_model.LinearRegression()

# Train the model
model.fit(X, y)

# Make a prediction for Cyprus
X_new = [[22587]]  # Cyprus' GDP per capita
print(model.predict(X_new)) # outputs [[6.1844]]
```

<font size=4> **_12 - What is the difference between a model parameter and a learning algorithm's hyperparameter?_**</font>

Um hiperparâmetro é um parâmetro do algorítimo de machine learning (e não do modelo) e tem a função de mensurar a _regularização_ a ser aplicada no modelo durante a fase de treinamento. * Parâmetros do modelo: θ_1 e θ_2, por exemplo;
* Hiperparâmetros: definem a regularização e não estão ligados com os parâmetros do modelo.

<font size=4> **_13 - What do model-based learning algorithms search for? What is the most common strategy they use to succeed? How do they make predictions?_**</font>

Os modelos que não são baseados em instância, são conhecidos como **Model-Based Learning** e, entre suas principais características, encontra-se o treinamento padrão em um conjunto de dados através de um modelo ou algorítimo de Machine Learning. Dessa forma, são definidos o tipo de algorítimo, são levantados os parâmetros do modelo, _cost function_ (mede eficiência do modelo) e uma série de outras tarefas bem conhecidas. A diferença para o método **Instance-Based Learning** é que, neste caso, as predições são realizadas de acordo como o modelo (por exemplo, uma regressão linear) e não através das instâncias/pontos já presentes (por exemplo, kNN). Pode haver relativa diferença entre estes modelos aplicados a um mesmo problema. Em resumo, um aprendizado baseado em modelo segue:
* Estudo dos dados;
* Seleção do modelo;
* Treinamento dos dados (levantamento de parâmetros que minimizam a _cost function_;
* Aplicação do modelo para predições (inferência).

<font size=4> **_14 - Can you name four of the main challenges in Machine Learning?_**</font>
* **Quantidade dos dados**
    - Para resolver um problema utilizando Machine Learning com eficiência, deve haver uma quantidade de dados razoável para que um modelo fiel possa ser aplicado, garantindo bons resultados através da generalização.
* **Representatidade e Qualidade dos dados**
     - Em outra vertente, uma grande massa de dados de baixa qualidade também não é válida. Os dados devem representar o comportamento real do problema a ser analisado. Paralelamente, um conjunto de treinamento com diversos _outliers_ ou ruídos dificultará o reconhecimento de padrões por parte do algorítimo, ocasionando em resultados que não demonstram a realidade.
* **Features irrelevantes**
    - É extremamente importante que haja uma avaliação das variáveis preditoras de maior importância para o modelo. Atributos ou features sem relevância para o processo acabam atuando de forma negativa nos resultados, ocupando espaços e interferindo na acurácia;
* **Overfitting**
    - Este é um grande desafio enfrentado durante o treinamento dos dados. A separação do conjunto de dados entre _treinamento_ e _teste_ é uma maneira de se alcançar melhores resultados no modelo, entretanto, a vaidade em se alcançar magníficos resultados com o conjunto de treinamento pode ocasionar no fenômeno conhecido como _overfitting_ que nada mais é do que a falta de generalização do modelo para dados que não façam parte do conjunto treinado. Em outras palavras, um modelo overfitting é perfeitamente capaz de reconhecer padrões presentes no conjunto de treinamento, porém extremamente incapaz de reconhecer qualquer dado externo. A redução em graus polinomiais é uma maneira de se evitar o overfitting.

<font size=4> **_15 - If your model performs great on the training data but generalizes poorly to new instances, what is happening? Can you name three possible solutions?_**</font>

Como mencionado acima, trata-se do fenômeno conhecido como **Overfitting**, onde o modelo é tão bem treinado com o conjunto de treinamento que passa a atuar de forma efetiva apenas em dados que contem este conjunto, mostrando resultados ruins com dados externos. Algumas soluções possíveis:
* Simplificação do modelo através da seleção de uma quantidade menor de parâmetros (theta), ou seja, aplicando uma redução no degrau polinomial da função (redução de atributos nos dados de treinamento).
* Coletando uma quantidade maior de dados de treinamento.
* Reduzindo ruídos no set de treinamento (removendo erros e outliers).

<font size=4> **_16 - What is a test set and why would you want to use it?_**</font>

Na fase de treinamento de um modelo de Machine Learning, normalmente o conjunto total de dados é dividido em dois subconjuntos: treinamento e teste. Isto é essencial para que uma esteira de testes seja implementada com eficiência, sendo possível checar se o treinamento foi realmente efetivo através da inserção de dados considerados externos. Com isso, é possível verificar se o modelo está generalizando bem e se os resultados estão plausíveis (isto pode ser feito através de medidas de acurácia). Normalmente utiliza-se 80% do conjunto total para treinamento e 20% para testes, porém este número pode variar dependendo do problema a ser resolvido.

<font size=4> **_17 - What is the purpose of a validation set?_**</font>

Um teste de validação é importante para que se seja possível certificar o nível de generalização do modelo de Machine Learning aplicado. Em outras palavras, é possível medir como o algoritimo se comporta ao receber dados externos aos dados de treinamento. Medidas de _acurácia_ são capazes de realizar este controle.

<font size=4> **_18 - What can go wrong if you tune hyperparameters using the test set?_**</font>

Ajustar os hiperparâmetros repetiedas vezes no set de treinamento, ocasiona uma má generalização do modelo, fazendo com que este apresente performances ruins quando apresentado a novos conjuntos de dados que não os de treinamento. Para sanar este tipo de problema, é comum criar um outro conjunto de dados chamado de _validation set_ ou então aplicar os conceitos de _cross validation_ que, por sua vez, nada mais é que a separação do conjunto de dados em partes complementares e, uma vez definidos os hiperparâmetros, o teste é realizado através do cruzamento destes pequenos conjuntos, onde cada um funciona como teste e treino pelo menos uma vez.

<font size=4> **_18 - What is a cross-validation and why would you prefer it to a validation set?_**</font>

Como mencionado na questão anterior, a validação cruzada (ou _cross validation_) é o processo de "quebrar" o dataset em pequenos conjuntos complementares, realizando o treinamento do modelo de modo a cruzar todos estes pequenos conjuntos, validando com a parte remanescente. É preferível realizar este processo do que separar um teste de validação pois assim os dados são aproveitados com mais eficiência.
