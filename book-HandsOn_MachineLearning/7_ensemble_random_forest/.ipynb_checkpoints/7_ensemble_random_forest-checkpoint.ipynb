## Introdução

Suponha que você faça uma pergunta complexa para milhares de pessoas aleatórias. Após isso, você agrega todas as respostas. Talvez você perceba que a resposta agregada vale mais que a resposta de um único expert. A isto dá-se o nome de _Wisdom of the Crowd_ (sabedoria do povo). Analogamente, se você agregar as **predições** de um grupo de **preditores**, provavalmente isto irá te levar a predições **melhores** que o melhor preditor analisado de forma individual. Um grupo de preditores é chamado _ensemble_ e, a esta técnica, dá-se o nome de _Ensemble Learning_. Um algoritmo de Ensemble Learning é chamado _Ensemble Method_.

Por exemplo, é possível treinar um grupo de classificadores _Decision Trees_, cada um em um subset aleatório do training set. Para realizar predições, basta obter as predições de cada um dos classificadores individualmente e então predizer a classe que ganhar mais votos. Um conjunto de _Decision Trees_ é chamado de _Random Forest_ e, apesar da aparente simplicidade, é um dos algoritmo mais poderosos de Machine Learning disponíveis hoje.

Os modelos Ensemble são geralmente discutidos no final de cada projeto, uma vez que a solução já foi construída e já foram testados e levantados bons modelos para solução do problema. Neste ponto, possível realizar combinações destes modelos com o objetivo de criar um modelo ainda mais poderosos para aquele problema em questão. Soluções vencedoras em competições de Machine Learning geralmente envolvem métodos Ensemble. Neste capítulo, _bagging, boosting, stacking_ e também _Random Forest_.

## Voting Classifiers

Suponha que você tenha treinado uma série de classificadores, cada qual obtendo 80% de acurácia. Você tem em mãos Regressão Logística, SVM, Random Forest, KNN e talvez alguns mais.

Um jeito simples de criar um classificador ainda melhor é agregar as predições de cada cassificador e retornar, como predição, a classe que ganhar a maior quantidade de votos. Esta classe mais votada é chamada de _hard voting_ classifier. Surpreendentemente, esse _voting classifier_ irá alcançar uma acurácia maior que o melhor classificador individual do conjunto. De fato, mesmo de um classificador é dado como um _weak learner_ (em outras palavras, algo próximo ao **aleatório**), o conjunto ainda pode ser um _strong learner_ (com uma boa **acurácia**), desde que haja um número suficiente de weak learner e que estes sejam suficientemente diversos.

**Como isso é possível?** - Pense na analogia da moeda.

Suponha que você tenha uma moeda viciada com 51% de chance de dar cara e 49% de dar coroa. Se você joga-la 1000 vezes, normalmente se espera um resultado 510 caras e 490 coroas ou, pelo menos, uma maioria de resultados em cara. Se você calcular matematicamente, verá que a probabilidade de obter maioria cara com esta moeda, após 1000 lançamentos, é de 75%. Quanto mais você joga a moeda, maior essa "probabilidade da maioria" (por exemplo, em 10.000 lançamentos, essa probabilidade sobre para 97%). 

Isto é devido a _law of large numbers_: quanto mais a moeda é lançada, mais a probabilidade de caras se assemelha ao vício da moeda (51%). Veja o gráfico abaixo onde isso foi testado para 10 séries diferentes. Em todas elas, há uma grande aproximação a linha tracejada que indica 51%, ou pelo menos as séries terminam acima de 50%.

```python
import numpy as np
heads_proba = 0.51
coin_tosses = (np.random.rand(10000, 10) < heads_proba).astype(np.int32)
cumulative_heads_ratio = np.cumsum(coin_tosses, axis=0) / np.arange(1, 10001).reshape(-1, 1)
```

```python
# To plot pretty figures
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt

plt.figure(figsize=(8,3.5))
plt.plot(cumulative_heads_ratio)
plt.plot([0, 10000], [0.51, 0.51], "k--", linewidth=2, label="51%")
plt.plot([0, 10000], [0.5, 0.5], "k-", label="50%")
plt.xlabel("Number of coin tosses")
plt.ylabel("Heads ratio")
plt.legend(loc="lower right")
plt.axis([0, 10000, 0.42, 0.58])
plt.show()
```

Analogamente, suponha um total de 1.000 **classificadores** que, individualmente, acertam em apenas 51% das oportunides, (próximo ao aleatório). Se predições forem realizadas considerando o voto da maioria, espera-se uma acurácia em torno de 75%! Contanto, isto somente é verdade caso os classificadores sejam totalmente independentes, com erros que não podem ser correlacionados, o que obviamente não é caso, dado que os modelos são treinados no mesmo conjunto de dados.
