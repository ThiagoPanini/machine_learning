## 3 - Recursos com valores faltantes
Referências: https://stackoverflow.com/questions/29530232/how-to-check-if-any-value-is-nan-in-a-pandas-dataframe

```python
# A função .isna(), se aplicado a um DataFrame inteiro, retorna um DataFrame de booleanos, o que não é muito útil neste caso.
df_red.isna()[:5]
```

```python
# Temos a função .isnull().values.any() que retorna True ou False dependendo se há ou não QUALQUER valor NaN

if df_red.isnull().values.any():
    print(f'Há valores NaN no Dataset de vinho tinto.')
else:
    print(f'Não há valores NaN no Dataset de vinho tinto.')
print()
if df_white.isnull().values.any():
    print(f'Há valores NaN no Dataset de vinho branco.')
else:
    print(f'Não há valores NaN no Dataset de vinho branco.')
```

```python
# Uma outra forma, um pouco mais lenta, mas também eficaz e específica, é utilizando a função .sum().sum()
# Com ela, é possível verificar com detalhes se há valores faltantes e sua quantidade.

print(f'Há um total de {df_red.isnull().sum().sum()} valores faltantes no Dataset de vinho tinto.')
print(f'Há um total de {df_white.isnull().sum().sum()} valores faltantes no Dataset de vinho branco.')
```

## 4 - Linhas duplicadas no conjunto de dados sobre vinho branco

Referências: https://medium.com/@kasiarachuta/dealing-with-duplicates-in-pandas-dataframe-789894a28911

```python
# A função .duplicated() retorna uma Series de valores booleanos dizendo True se a linha for duplicada e False caso contrário.
print(df_white.duplicated()[:5])
print(type(df_white.duplicated()))
```

```python
# Para contabilizar as linhas duplicadas, podemos usar .sum() nos valores booleanos (soma apenas True)]
print(f'Há um total de {df_white.duplicated().sum()} linhas duplicadas neste Dataset')
```

```python
# Passando uma Series como argumento para indexar o Dataset, é possível criar um novo DataFrame apenas com as duplicadas
df_white_duplicated = df_white[df_white.duplicated()]
df_white_duplicated.head()
```

A conclusão que se chega sobre as linhas duplicadas é que não necessariamente elas precisam ser apagadas. Isto pois, como se tratam de atributos químicos, é perfeitamente possível duas ou mais amostras possuírem quantidades iguais dentro destes atributos. Dessa forma, o mais adequado é manter estes dados duplicados.

Objetivo:
* Criar uma coluna 'color' em cada um dos dataframes 
* Atribuir os respectivos valores, ou seja, 'red' para df_red e 'white' para df_white
* Juntar os dois DataFrames

```python
# Importando biblioteca numpy 
import numpy as np
```

```python
# Utilizando objeto do tipo array para criar nova coluna de dados
color_red = np.repeat(np.array(['red']), df_red.shape[0])
color_white = np.repeat(np.array(['white']), df_white.shape[0])
```

```python
# Cada um dos comandos acima cria um array unidimensional com valores repetidos de 'red' e 'white'
color_red[:5]
```

```python
color_white[:5]
```

```python
# Verificando comprimento de cada um
print(color_red.shape)
print(color_white.shape)
```

```python
# Criando nova coluna 'color' em cada um dos DataFrames
df_red['color'] = color_red
df_white['color'] = color_white
```
