## Introdução

O *Apache Hadoop* é um software open source para armazenamento e processamento em larga escala (Big Data) em *clusters* de hardware de baixo custo. Mas afinal, qual a definição de *cluster*?

**Cluster:** conjunto de computadores que trabalham em conjunto e se comportam como se fossem apenas um. Esse procedimento aumenta a capacidade de armazenamento e processamento de forma exponencial!

Com isso em mente, é possível entender o Hadoop como um sistema de armazenamento compartilhado, distribuído e altamente confiável para processamento de grandes volumes de dados através de clusters de computadores. Com essa filosofia, é possível utilizar o HD, a memória e a CPU de várias máquinas ao mesmo tempo, aumento a capacidade de maneira extraordinária.

**O Hadoop nasceu na era do Big Data!**

### Hadoop e Big Data

O Apache Hadoop foi criado para responder as principais questões propostas pelo Big Data e resolver seus principais problemas. Por este motivo, os Bancos de Dados Relacionais apresentam grandes dificuldades a medida que estamos próximos a problemas de Big Data.

<img src="images\02_hadoop_cicle.png" alt="hadoop_cycle" style="width: 400px;"/>

Dessa forma, é possível encontrar no Hadoop, soluções próprias para os problemas trazidos pelo Big Data, como por exemplo, o armazenamento e processamento de dados em grandes volumes (fatores já citados em larga escala).

### Principais Módulos

O Apache Hadoop é composto de 3 módulos principais:

* _Hadoop Distributed File System (HDFS)_

Se há diversas máquinas funcionando como apenas uma (clusters), é desejável utilizar o HD de cada uma delas como um único grande disco. O HDFS é um software para gerenciar os discos dessas diversas máquinas para que estas se comportem efetivamente como uma única máquina.

Quando o Cientista de Dados realiza a carga de dados no Hadoop, o pacote é enviado ao HDFS que, por sua vez, é encarregado pela distribuição dos dados ao longo de vários HDs em várias máquinas diferentes. Isto serve tanto para gravação quanto para a leitura dos dados.

* _Hadoop Yarn_

O objetivo do Yarn é realizar a **gestão** de todo esse processo. Trata-se de um componente mais recente que os demais.

* _Hadoop MapReduce_

O objetivo do Map Reduce é o de gerenciar o **processamento** dos dados de forma paralela. Enquanto o HDFS é responsável pelo armazenamento, o Map Recude trata da forma como que estes dados são processados. Para contextualizar, há o **Apache Spark** que, por sua vez, roda sobre o HDFS e é uma alternativa mais eficiente para o processamento dos dados.

### Inspiração

O Hadoop é um framework gratuito baseado na linguagem de programação Java. Como inspiração, foi baseado no **Google File System (GFS)**.

Com o Hadoop, é possível executar aplicações em sistemas distribuídos através de **nodes** envolvendo **petabytes** de dados. Seria impossível realizar procedimentos semelhantes em uma máquina comum. Para isso, são necessários clusters. Para a criação e gerenciamento de clusters, é necessário um software específico. Este software é o Hadoop!

* Gerenciamento de Problemas

Mesmo que um node saia de circulação ou apresente algum problema, o HDFS garante a recuperação a falha e o não comprometimento do procedimento executado.

## Hadoop HDFS

<img src="images\05_hadoop_hdfs.png" alt="hadoop_hdfs" style="width: 300px;"/>

Algumas características do HDFS:
    * Tolerância a falhas e recuperação automática;
    * Portabilidade entre hardware e sistemas operacionais heterogêneos;
    * Escalabilidade para armazenar e processar grandes quantidades de dados;
    * Confiabilidade, através da manutenção de várias cópias de dados.

HDFS - Hadoop Distributed File System
    - Sistema de Arquivos Distribuídos do Hadoop

A tolerância a falhas é uma das principais características do HDFS. Se uma máquina apresentar defeito dentro de um cluster de 30 máquinas, por exemplo, será possível continuar o armazenamento (ou o processamento) dos dados sem que o cluster como um todo seja atingido.

Um outro ponto importante é que o HDFS foi pensado para ser ótimo em performance do tipo WORM (Write Once Read Many), que é um eficiente padrão de processamento de dados.

### Nodes

O HDFS possui 2 tipos de nodes:
* Namenode (masternode)
* Datanode (worker node)

<img src="images\07_hadoop_hdfsnodes.png" alt="hadoop_hdfsnodes" style="width: 550px;"/>

**Namenode:**
    - Gerencia a estrutura o filesystem;
    - Gerencia os metadados de todos os arquivos e diretórios dentro da estrutura;
    - É recomendado que o namenode seja uma máquina com maior capacidade computacional;


**Datanode:**
    - Armazena e busca blocos de dados quando solicitado pelo cliente ou pelo Namenode;
    - Trabalha com cada um dos blocos de dados para armazenamento e leitura;
    - Reporta periodicamente para o Namenode com a lista de blocos que foram armazenados.
    - Este report é fundamental para que o Namenode possa gerenciar o aramzenamento.

No curso *Engenharia de Dados com Hadoop e Spark*, é possível aprender a instalar o Namenode, configurar o Datanote, fazer com que os dois se comuniquem e configurar todo o procedimento para armazenamento de dados. Na *Formação Engenheiro de Dados*, é proposta a criação de um Data Lake completo, integrando o Hadoop com outras soluções, como por exemplo, Banco de Dados No SQL, Apache Spark, entre outros.

## Hadoop MapReduce

<img src="images\06_hadoop_mapreduce.png" alt="hadoop_hdfs" style="width: 300px;"/>

Algumas características do MapReduce:
    * Flexibilidade - processa os dados independente do tipo e formato, estruturado ou não;
    * Confiabilidade - jobs são executados em paralelo e não são afetados em caso de falhas;
    * Acessibilidade - suporte a diversas linguagens como Java, C++, Python

Uma vez que os dados são armazenados e gerenciados com o HDFS, é necessário processar estes dados, extrair informações úteis, aplicar modelos de Machine Learning, realizar análises e buscar padrões. Isto é feito a partir do MapReduce.

`MapReduce` é um modelo de programação de processamento de grandes conjuntos de dados. **É importante ressaltar que o conceito de MapReduce existe independentemente do Hadoop.** É possível realizar o mapeamento e redução utilizando outras ferramentas ou outros framewoorks.

* MapReduce transforma o problema de análise em um processo computacional que usa conjuntos de chaves e valores;
* MapReduce foi desenvolvido para tarefas que consomem minutos ou horas em computadores conectados em rede de alta velocidade gerenciados por um único master.
* MapReduce usa um tipo de análise de dados por força bruta. Todo o conjunto de dados é processado em cada query.
* MapReduce usa um modelo de processamento em batch, ao contrário de um modelo de processamento contínuo (será abordado nos tópicos referentes ao `Apache Spark`)

### Chaves e Valores

A ideia do `Mapeamento` e `Redução` gira em torno da aplicação de uma regra de mapeamento em um conjunto de dados, de acordo com o problema a se resolver com o procedimento de análise, gerando assim vários conjuntos de _chaves_ e _valores_.

<img src="images\08_hadoop_mrkv.png" alt="hadoop_hdfs" style="width: 300px;"/>

Em linhas gerais, o Cientista de Dados deve identificar o problema a ser resolvido, definir e aplicar alguma técnica específica e, assim, criar o procedimento de Mapeamento e Redução que, por sua vez, pode alimentar um outro sistema de Machine Learning, um sistema de Recomendação e assim por diante.

<img src="images\09_hadoop_mr.png" alt="hadoop_hdfs" style="width: 500px;"/>

Por que o modelo de MapReduce é ideal para o processamento de Big Data?

### Seek Time e Transfer Rate

MapReduce:
    - Execução de queries ad-hoc (direto no cluster, sem interface ou sistema intermediário)
    - Disparo de queries ad-hoc via terminal;
    - Performance relacionada ao balanceamento entre seeking e transfer.

**Ideia:** reduzir operações seeking e usar de forma efetiva as operações de transfer.

**Não adianta ter um armazenamento paralelo e um tempo de processamento altíssimo!**

`Seek Time:` delay para encontrar um arquivo;

`Transfer Rate:` velocidade para encontrar o arquivo (vem melhorando significativamente).

No MapReduce, todo (ou a maior parte) do conjunto de dados é atualizado. No RDBMS (Relation Database Management System), é bom para atualizar pequenas porções de grandes bancos de dados.

`RDBMS:` Utiliza B-Tree, que é altamente dependente de operações seek;

`MapReduce:` Utiliza operações de SORT e Merge, mais dependente de operações transfer.
