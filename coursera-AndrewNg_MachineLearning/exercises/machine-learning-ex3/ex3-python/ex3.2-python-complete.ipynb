Este notebook tem por objetivo resolver o mesmo problema de classificação de dígitos exposto no notebook ```3.1```, com a diferença no algoritmo utilizado: ```Redes Neurais Artificiais```. Disposto do mesmo dataset de dígitos manuscritos, serão aplicadas funções específicas deste tipo de algoritmo, como ```forward propagation``` e ```back propagation``` para se chegar a resultados efetivos.

A grande vantagem de treinar um modelo de Redes Neurais, em comparação com um modelo de Regressão Logística é que, neste segundo, há uma certa dificuldade em implementar funções não lineares, visto que o peso computacional é elevado para funções de alto grau polinomial e com uma gigantesca presença de features (um feature mapping em um modelo de Regressão Logística, para uma dataset de imagens com 400 features, poderia gerar milhares de features adicionais).

## Redes Neurais Artificiais

### Arquitetura da Rede

Para a resolução do problema, foi proposta a seguinte rede:

Temos uma camada de entrada, uma camada oculta e uma camada de saída.

Uma informação extremamente importante: **os parâmetros theta da rede já foram previamente estabelecidos**.

### Lendo Dados

````python
# Importando biblioteca
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
# Definindo função para leitura dos dados
def load_data():
    """ Função responsável por ler os dados e a matriz de parâmetros
    Output:
        X: features do modelo
        y: vetor de label (10 classes)
        Theta1: matriz de parâmetros da primeira camada da rede
        Theta2: matriz de parâmetros da segunda camada da rede"""
    
    # Lendo dados
    data = sio.loadmat('ex3data1.mat')
    X = data['X']
    y = data['y']
    
    params = sio.loadmat('ex3weights.mat')
    Theta1 = params['Theta1']
    Theta2 = params['Theta2']
    
    return X, y, Theta1, Theta2
````

````python
# Lendo arquivos
X, y, theta1, theta2 = load_data()
````

````python
# Verificando dimensões
y = y.reshape(-1)
print(f'Dimensões de X: {X.shape}')
print(f'Dimensões de y: {y.shape}')
print(f'Dimensões de Theta1: {theta1.shape}')
print(f'Dimensões de Theta2: {theta2.shape}')
````

**Explicação sobre as dimensões dos parâmetros theta:**

Como regra das Redes Neurais, temos que as dimesões dos parâmetros que relacionados a camada ```j``` são dadas como a quantidade ```sl``` de unidades da camada ```j+1``` _versus_ a quantidade ```sl``` de unidades da camada ```j``` + 1.

**Exemplo prático:**

Em uma rede com **2** nós de entrada e uma segunda camada com **4** nós, temos que as dimensões da matriz de parâmetros theta1 seria dada por ```4``` (sl de j+1), ```3``` (sl de j + 1), resultando em theta1 = (4, 3).

**Parâmetros para este caso:**

**Theta1**: temos 400 nós (features -> imagens 20x20) na camada de entrada (contando com a unidade de bias) e 25 nós na camada oculta. Como este parâmetro define o comportamento da rede entre a camada de entrada e a camada oculta, temos que suas dimensões são ```(25, 401)```.

**Theta2**: como dito acima, temos 25 nós na camada oculta e, como informação nova, temos 10 nós na camada de saída (labels -> 10 dígitos). Como este parâmetro define o comportamento da rede entre a camada oculta e a camada de saída, temos que suas dimensões são ```(10, 26)```.

````python
# Definindo função para plotagem aleatório de um dígito
def plot_digit(X, y):
    """ Função responsável pela plotagem de um dígito do dataset de forma aleatória
    Input:
        X: features do modelo;
        y: vetor de labels"""
    
    m = X.shape[0]
    rnd = np.random.randint(m)
    plt.imshow(X[rnd].reshape(20, 20).T, cmap='viridis')
    plt.axis('off')
    plt.title(f'Dígito: {y[rnd]}', size=13)
    plt.show()
````

### Dados de Treino e Teste

````python
# Definindo função para splitar os dados
def shuffle_and_split_data(X, y, test_ratio):
    """ Função responsável por realizar o embaralhamento nos dados e separação
    em conjuntos de teste e treino
    Input:
        X: features do modelo
        y: vetor de label
        test_ratio: porcentagem dos dados no conjunto de teste
    Output:
        X_train, X_test, y_train, y_test: conjuntos de dados splitados"""
    
    # Criando índices embaralhados
    m = X.shape[0]
    idx = np.random.RandomState(seed=42).permutation(m)
    X_shuffle = X[idx]
    y_shuffle = y[idx]
    
    # Separando os dados
    test_indices = int(m * test_ratio)
    X_train = X_shuffle[test_indices:]
    X_test = X_shuffle[:test_indices]
    y_train = y_shuffle[test_indices:]
    y_test = y_shuffle[:test_indices]
    
    # Retorno da função
    return X_train, X_test, y_train, y_test
````

````python
# Lendo dados e parâmetros
X, y, Theta1, Theta2 = load_data()
````

````python
# Separando dados
X_train, X_test, y_train, y_test = shuffle_and_split_data(X, y, .20)
````

````python
# Verificando
y_train = y_train.reshape(-1)
y_test = y_test.reshape(-1)
print('- - - Treino - - -')
print(f'Dimensões de X_train: {X_train.shape}')
print(f'Dimensões de y_train: {y_train.shape}')
print()
print('- - - Teste - - -')
print(f'Dimensões de X_test: {X_test.shape}')
print(f'Dimensões de y_test: {y_test.shape}')
````

````python
# Parâmetros
print(f'Dimensões de Theta1: {Theta1.shape}')
print(f'Dimensões de Theta2: {Theta2.shape}')
````

````python
# Definindo função para cálculo da função h
def hypothesis(X, Theta1, Theta2):
    """ Função responsável por cálcular vetor de predições de acordo com os dados inputados
    Input:
        X: dados de dimensões m x n inseridos para cálculo de predições
    Output:
        h = vetor m x 10 com predições relacionadas aos dados"""
    
    # Adicionando termo de bias nos dados
    m = X.shape[0]
    X = np.c_[(np.ones((m, 1)), X)]
    
    # Calculando função da camada 2
    a2 = sigmoid(X.dot(Theta1.T))
    a2 = np.c_[(np.ones((m, 1)), a2)]
    
    # Calculando função da camada de saída
    h = sigmoid(a2.dot(Theta2.T))
    
    return h
````

````python
# Verificando
h = hypothesis(X_test, Theta1, Theta2)
print(f'Dimensões da matriz de predições: {h.shape}')
````

````python
# Definindo função para retornar a predição de cada linha
def predictions(h, y):
    """ Função responsável por cálcular as predições e retornar a acurácia do modelo
    Input:
        h: matriz de probabilidades de dimensões m x k
        y: vetor de labels de dimensões m x 1
    Output:
        acc: acurácia do modelo"""
    
    # Configurando parâmetros
    m = h.shape[0]
    pred = np.zeros((m, 1)).reshape(-1)
    
    # Laço para cálculo das predições
    for i in range(m):
        pred[i] = np.where(h[i] == h[i].max())[0] + 1
    
    # Calculando acurácia
    acertos = sum((pred == y) * 1)
    return acertos / m
````

````python
# Iniciando função
acc = predictions(h, y_test)
print(f'Acurácia do modelo: {acc:.2f}')
````

Ainda sim, muito bom! Os parâmetros pré-configurados estavam muito bem calibrados!

## Scikit Learn

No exercício acima, o set de parâmetros já estava previamente carregado. Agora, vamos utilizar o treinamento de redes neurais proposto pelo ```scikit learn``` para construir um modelo de reconhecimento de dígitos tão bom quanto o já fornecido.

````python
# Importando biblioteca
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
# Definindo função para leitura dos dados
def load_data():
    """ Função responsável por ler os dados e a matriz de parâmetros
    Output:
        X: features do modelo
        y: vetor de label (10 classes)"""
    
    # Lendo dados
    data = sio.loadmat('ex3data1.mat')
    X = data['X']
    y = data['y']
    
    return X, y.reshape(-1)

# Definindo função para plotagem aleatório de um dígito
def plot_digit(X, y):
    """ Função responsável pela plotagem de um dígito do dataset de forma aleatória
    Input:
        X: features do modelo;
        y: vetor de labels"""
    
    m = X.shape[0]
    rnd = np.random.randint(m)
    plt.imshow(X[rnd].reshape(20, 20).T, cmap='viridis')
    plt.axis('off')
    plt.title(f'Dígito: {y[rnd]}', size=13)
    plt.show()
````

### Aumentando Dataset

Vamos agora aplicar um procedimento de ```data augmentation``` para deslocar cada uma das imagens do nosso dataset, aumentando assim o nosso conjunto de dados e ajudando o modelo a melhorar sua perfomance. Vejamos.

````python
from scipy.ndimage.interpolation import shift

def shift_image(image, dx, dy):
    """
    Função responsável por deslocar imagens do dataset em determinada direção
    
    Input: 
        image (registro do dataset X_train)
        dx, dy (deslocamentos em x e y)
    Output:
        shifted_image (imagem deslocada)
    """
    image = image.reshape((20, 20))
    shift_image = shift(image, [dy, dx], cval=0, mode='constant')
    return shift_image.reshape([-1])

def augment_data(X_train, y_train):
    """
    Função para deslocar imagens do dataset em todas as direções
    
    Input:
        X_train, y_train (dados brutos de entrada)
    Output:
        X_train_augmented, y_train_augmented (dados deslocados e embaralhados em formato de array)
    """
    X_train_augmented = [image for image in X_train]
    y_train_augmented = [label for label in y_train]
    
    for dx, dy in ((1, 0), (-1, 0), (0, 1), (0, -1)):
        for image, label in zip(X_train, y_train):
            X_train_augmented.append(shift_image(image, dx, dy))
            y_train_augmented.append(label)
    
    X_train_augmented = np.array(X_train_augmented)
    y_train_augmented = np.array(y_train_augmented)
    
    shuffle_idx = np.random.permutation(len(X_train_augmented))
    X_train_augmented = X_train_augmented[shuffle_idx]
    y_train_augmented = y_train_augmented[shuffle_idx]
    
    return X_train_augmented, y_train_augmented

def plot_data_augmented(X):
    """
    Função responsável por plotar graficamente imagens deslocadas do dataset
    """
    image = X[np.random.randint(0, X.shape[0])]
    shifted_image_down = shift_image(image, 0, 5)
    shifted_image_left = shift_image(image, -5, 0)
    shifted_image_up = shift_image(image, 0, -5)
    shifted_image_right = shift_image(image, 5, 0)
    
    # Criando figure e plotando gráficos
    plt.figure(figsize=(12, 3))
    plt.subplot(151)
    plt.title('Original', fontsize=14)
    plt.imshow(image.reshape(20, 20).T, interpolation="nearest", cmap="Greys")
    plt.subplot(152)
    plt.title('Shifted down', fontsize=14)
    plt.imshow(shifted_image_down.reshape(20, 20).T, interpolation="nearest", cmap="Greys")
    plt.subplot(153)
    plt.title('Shifted left', fontsize=14)
    plt.imshow(shifted_image_left.reshape(20, 20).T, interpolation="nearest", cmap="Greys")
    plt.subplot(154)
    plt.title('Shifted up', fontsize=14)
    plt.imshow(shifted_image_up.reshape(20, 20).T, interpolation="nearest", cmap="Greys")
    plt.subplot(155)
    plt.title('Shifted right', fontsize=14)
    plt.imshow(shifted_image_right.reshape(20, 20).T, interpolation="nearest", cmap="Greys")
    
    plt.show()
````

````python
# Retornando dados aprimorados
X_train_augmented, y_train_augmented = augment_data(X_train, y_train)
````
